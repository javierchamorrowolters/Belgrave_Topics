## Titanic

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
pd.set_option("display.max_columns", 1000)
```

```python
pd.get_dummies(titanic)
```

### 1.Get mean survivers by sex

```python
titanic[['survived','sex']].groupby('sex').mean()
```

###Â 2. Get the mean number of survivers by sex and by class. User Class in the column

```python
titanic[['survived','sex', 'class']].groupby(['sex', 'class']).mean()
```

###Â 3. Get the total number of survivers by sex and class. And the average fare for each of those groups

```python
titanic[['survived','sex', 'class']].groupby(['sex', 'class']).sum()
```

```python
titanic[['survived','sex', 'class', "fare"]].groupby(["survived", 'sex', 'class']).mean()
```

###Â 4. Get the total number of survivers per embark_town.

```python
titanic[['survived',"embark_town"]].groupby('embark_town').sum()
```

###Â 5. Get the percentage of survivers per embark_town.

```python
titanic[['survived',"embark_town"]].groupby('embark_town').mean()*100
```

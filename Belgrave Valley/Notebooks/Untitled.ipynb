```python
'''Create a function that counts all 9s in a range'''
def nine_count(from_, to_):
    count_of_nines = 0
    for x in list(range(from_,to_+1)):
        for letter in str(x):
            if letter == str(9):
                count_of_nines += 1
    return count_of_nines
        
```

```python
nine_count(0,1000)
```

```python
def nine_count_tq(from_, to_):
    count_of_nines = 0
    for x in tqdm.tqdm(list(range(from_,to_+1))):
        for letter in str(x):
            if letter == str(9):
                count_of_nines += 1
    return count_of_nines
```

```python
list_3 = [1,1,1,3,3,2,5,6]
```

```python
def freq_t(lis):
    dic = {}
    for x in lis:
        if x in dic.keys():
            dic[x] += 1
        else:
            dic[x] = 1
    return dic
```

```python
j = freq_t(list_3)
j
```

```python
plt.bar(x=j.keys(), height=j.values())
```

```python
def pmf(diction):
    pmf = {}
    suma_total = 0
    for x in diction.values():
        suma_total += x
    
    for x in diction.keys():
        pmf[x] = (diction[x]/suma_total)
    
    return pmf
```

```python
i = pmf(j)
i
```
